```python
# 엑셀파일 읽기

df = pd.read_excel("경기도인구데이터.xlsx")
df[:3]
```

```python
# 칼럼 이름 문자형으로 바꾸기

df.columns = df.columns.astype('str')

# df.columns = df.columns.map(str)
# df.columns = map(str, df.columns)
# df.columns = map(lambda x : str(x), df.columns)
```

```python
df.set_index('구분', drop = False, inplace = True)
df
```

```python
# json 파일 일기

geo_data = json.load(open("경기도행정구역경계.json", encoding = 'utf-8-sig'))   # utf-8로 못 읽을 때 utf-8-sig 사용
```

```python
# 위도 27.55, 경도 126.98

g_map = folium.Map(location=[37.55,126.98], zoom_start=9)
```

```python
folium.Choropleth(geo_data = geo_data,   # 지도 경계
                 data = df['2017'],    # 표시하려는 데이터
                 columns = [df.index, df['2017']],  # 열 지정
                 fill_color = 'YlOrRd',
                 fill_opacity = 0.7,
                 line_opacity = 0.3,
                 threshold_scale = [10000, 200000, 300000, 500000, 700000],
                 key_on = 'feature.properties.name').add_to(g_map)    # 앞에 지도에 경계선 추가
```
